О ГОСТ

Стандарт ГОСТ 28147-89 регламентирует аспекты симметричного блочного шифрования, а также предписывает возможные режимы его использования. Относится к DES-подобным криптосистемам, при шифровании применяется схема Фейстеля. Разработка этого стандарта шифрования стартовала в 1976 году, а нормативный акт вступил в силу 1 июля 1990 года.

Средства криптографической защиты информации, сертифицированные в РФ, обозначаются аббревиатурой СКЗИ. С 1 июня 2019 года поддержка ГОСТ 28147-89 в новых СКЗИ не осуществляется. Исключение составляют случаи, когда на установленном ранее оборудовании уже использован этот алгоритм и нужно создать совместимое изделие.


Выдачей сертификатов соответствия для средств криптографической защиты информации в нашей стране занимается ФСБ России. Чтобы пройти сертификацию, обязательно применение отечественных алгоритмов.


Варианты алгоритмов для шифрования

Шифр блочного типа советской разработки известен как алгоритм криптографического преобразования ГОСТ 28147-89. Информация делится на модули размерностью 64 бита, каждый зашифровывается индивидуально.


Система работает следующим образом:

Сформированные изначально блоки делятся на две части.
На выходе данные с правого сегмента вписываются в левую сторону.
Правая часть преобразуется криптографическим способом с использованием ключа.
Левый сегмент с исходником и правый с преобразованными данными соединяются с помощью сумматора по второму модулю.
Система повторяет процесс еще несколько раз для повышения уровня безопасности.

Криптографический алгоритм ГОСТ 28147-89 по уровню надежности не уступает большинству зарубежных разработок.

Регламенты функционирования ГОСТ 28147-89

Алгоритм шифрования ГОСТ 28147-89 может функционировать в четырех вариантах:

Простая замена. Информация преобразовывается за счет распараллеливания на нужное число потоков. Этот вариант отличает высокая производительность. Но есть минус – одни и те же блоки открытых данных преобразуются в идентичные зашифрованные сегменты. Это позволяет раскрыть его структуру. Не имеет смысла шифровать таким методом больше, чем 1 блок текста.

Режим гаммирования предусматривает дробление данных на блоки «s». Впоследствии исходные сегменты соединяются с усеченными зашифрованными (состоят из синхропосылки IV и определенного значения счетчика). Этот вариант используется чаще всего, так как обеспечивает высокую скорость. В этом случае применяется способ прямого криптографического преобразования Ek.

Режим гаммирования дополненный обратной связью в ГОСТ 28147-89 представляет собой особый метод применения симметричного шифра в блочном исполнении. При таком подходе любой последующий блок суммируется по модулю 2 с перешифрованным предыдущим сегментом. При возникновении ошибок передачи не получится расшифровать только поврежденный элемент, на остальные это никак не влияет.

Режим выработки имитовставки. Процесс шифрования подобен простой замене с зацеплением. Технология отличается за счет формирования последнего блока, где используется производный ключ, который меняется в зависимости от того, был последний сегмент с исходной информацией полным, или нет.

```python
# Импорт библиотеки requests для выполнения HTTP-запросов
import requests
# Импорт BeautifulSoup из библиотеки bs4 для парсинга HTML
from bs4 import BeautifulSoup
# Импорт webdriver из selenium для автоматизации браузера
from selenium import webdriver
# Импорт Options для настройки параметров браузера Chrome
from selenium.webdriver.chrome.options import Options
# Импорт pandas для работы с данными в табличном формате (DataFrame)
import pandas as pd
# Импорт time для работы с задержками (sleep)
import time
# Импорт re для работы с регулярными выражениями
import re
# Импорт numpy для математических операций
import numpy as np
# Импорт matplotlib.pyplot для визуализации данных
import matplotlib.pyplot as plt
# Импорт seaborn для улучшенной визуализации данных
import seaborn as sns
# Импорт WordCloud для создания облака слов
from wordcloud import WordCloud
# Импорт Counter для подсчета частоты элементов
from collections import Counter
# Импорт urljoin для корректного объединения URL
from urllib.parse import urljoin

# Класс BookSpider для парсинга данных о книгах
class BookSpider:
    # Инициализация класса с базовым URL, максимальным количеством страниц и режимом headless
    def __init__(self, base_url, max_pages=3, headless=True):
        self.base_url = base_url  # Базовый URL сайта
        self.max_pages = max_pages  # Максимальное количество страниц для парсинга
        self.data = []  # Список для хранения данных о книгах
        self.setup_selenium(headless)  # Настройка Selenium
        self.visited_urls = set()  # Множество для отслеживания посещенных URL

    # Метод для настройки Selenium WebDriver
    def setup_selenium(self, headless):
        options = Options()  # Создание объекта для настройки опций Chrome
        if headless:
            options.add_argument('--headless')  # Включение headless-режима (без графического интерфейса)
        options.add_argument('--no-sandbox')  # Отключение sandbox для работы в некоторых средах
        options.add_argument('--disable-dev-shm-usage')  # Отключение /dev/shm для избежания ошибок
        self.driver = webdriver.Chrome(options=options)  # Инициализация драйвера Chrome с заданными опциями

    # Метод для загрузки страницы
    def fetch_page(self, url):
        try:
            if url in self.visited_urls:  # Проверка, посещалась ли страница ранее
                return None

            self.driver.get(url)  # Загрузка страницы в браузере
            time.sleep(np.random.uniform(1, 3))  # Случайная задержка между 1 и 3 секундами
            self.visited_urls.add(url)  # Добавление URL в множество посещенных
            return self.driver.page_source  # Возврат HTML-кода страницы
        except Exception as e:
            print(f"Error fetching {url}: {e}")  # Вывод ошибки, если загрузка не удалась
            return None

    # Метод для парсинга HTML-страницы
    def parse_page(self, html):
        soup = BeautifulSoup(html, 'html.parser')  # Создание объекта BeautifulSoup для парсинга
        books = soup.find_all('article', class_='product_pod')  # Поиск всех элементов с книгами

        page_data = []  # Список для хранения данных о книгах на текущей странице
        for book in books:  # Перебор всех найденных книг
            try:
                title = book.h3.a['title']  # Извлечение названия книги
                price = float(book.find('p', class_='price_color').text[1:])  # Извлечение цены (без символа £)
                rating = book.p['class'][1]  # Извлечение рейтинга (например, 'Three' → 3)
                rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}  # Словарь для преобразования рейтинга
                availability = book.find('p', class_='instock').text.strip()  # Извлечение информации о наличии

                page_data.append({  # Добавление данных о книге в список
                    'title': title,
                    'price': price,
                    'rating': rating_map.get(rating, 0),  # Преобразование рейтинга в число
                    'availability': availability,
                    'timestamp': pd.Timestamp.now()  # Текущая дата и время
                })
            except Exception as e:
                print(f"Error parsing book: {e}")  # Вывод ошибки, если парсинг не удался
                continue

        # Поиск ссылки на следующую страницу
        next_page = soup.find('li', class_='next')
        if next_page:
            next_url = urljoin(self.base_url, next_page.a['href'])  # Объединение базового URL и относительной ссылки
            return page_data, next_url  # Возврат данных и URL следующей страницы
        return page_data, None  # Возврат данных, если следующей страницы нет

    # Основной метод для запуска парсинга
    def crawl(self):
        current_url = self.base_url  # Начальный URL
        page_count = 0  # Счетчик посещенных страниц

        while current_url and page_count < self.max_pages:  # Цикл, пока есть страницы и не достигнут лимит
            print(f"Crawling {current_url}")  # Вывод текущего URL
            html = self.fetch_page(current_url)  # Загрузка страницы
            if html:
                page_data, next_url = self.parse_page(html)  # Парсинг страницы
                self.data.extend(page_data)  # Добавление данных в общий список
                current_url = next_url  # Переход на следующую страницу
                page_count += 1  # Увеличение счетчика страниц

    # Метод для сохранения данных в разных форматах
    def save_data(self, format='csv', filename='books_data'):
        df = pd.DataFrame(self.data)  # Создание DataFrame из собранных данных
        if format == 'csv':
            df.to_csv(f"{filename}.csv", index=False)  # Сохранение в CSV
        elif format == 'json':
            df.to_json(f"{filename}.json", orient='records')  # Сохранение в JSON
        elif format == 'excel':
            df.to_excel(f"{filename}.xlsx", index=False)  # Сохранение в Excel
        print(f"Data saved to {filename}.{format}")  # Подтверждение сохранения
        return df  # Возврат DataFrame

    # Метод для закрытия драйвера браузера
    def close(self):
        self.driver.quit()  # Завершение работы браузера
```

```python
# Создание экземпляра класса BookSpider
spider = BookSpider(
    base_url="https://books.toscrape.com",
    max_pages=5,  # Увеличим количество страниц для большего датасета
    headless=True # - headless=True: браузер будет работать в фоновом режиме без графического интерфейса
)

# Запуск процесса парсинга:
# - последовательный обход страниц сайта
# - сбор информации о книгах
# - сохранение данных в атрибуте spider.data
spider.crawl()

# Сохранение собранных данных в CSV-файл:
# - format='csv': выбор формата CSV
# - filename='books_data': имя файла без расширения
# - результат сохраняется в переменную df (DataFrame pandas)
df = spider.save_data(format='csv', filename='books_data')

# Завершение работы:
# - закрытие браузера
# - освобождение ресурсов
spider.close()

# Просмотр первых 5 строк полученных данных:
# - df.head() выводит "шапку" DataFrame
# - позволяет быстро проверить корректность собранной информации
df.head()
```
